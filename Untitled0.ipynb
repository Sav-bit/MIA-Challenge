```python
from ipywidgets import Text, Button, VBox, HTML
from IPython.display import display

name_input = Text(
    description='Sender:',
    placeholder='Type the sender name'
)

ok_btn = Button(description='OK')
status = HTML()

def on_click(b):
    global sender_name
    sender_name = name_input.value
    status.value = f"Sender name set to: <b>{sender_name}</b>"

ok_btn.on_click(on_click)

ui = VBox([name_input, ok_btn, status])
display(ui)
```
